```python
import pandas as pd
import numpy as np
```

```python
# Import the cleaned Data from our Repo Github
url ="https://raw.githubusercontent.com/ajdelcarpiogom/FinalProject/main/cleaned_data_andrea.csv"

df = pd.read_csv(url,index_col=0,parse_dates=[0])

df.head(5)
```

```python

df=df[df!=0].dropna()

df
```

```python
Sum_new_cases_per_countries = df.groupby(["location"]).sum()["new_cases"]
Mean_new_cases_per_countries = df.groupby(["location"]).mean()["new_cases"]
Sum_new_death_per_countries = df.groupby(["location"]).sum()["new_deaths"]
Mean_new_death_per_countries = df.groupby(["location"]).mean()["new_deaths"]
```

```python
figures_by_countries = pd.DataFrame({
               "Sum_new_cases_per_countries": Sum_new_cases_per_countries,
               "Mean_new_cases_per_countries": Mean_new_cases_per_countries,
               "Sum_new_death_per_countries": Sum_new_death_per_countries,
               "Mean_new_death_per_countries": Mean_new_death_per_countries
               })
figures_by_countries
```

```python
# sort Brand in an ascending order
figures_by_countries.sort_values(by=['Sum_new_death_per_countries'], inplace=True)
figures_by_countries.tail()
```
